```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Question 3: Normal Density Function and Numerical Integration
def normal_density(mean, variance, x):
    """Computes the normal density function."""
    sigma = np.sqrt(variance)
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean) ** 2) / (2 * variance))

# Plot normal density for different means and variances
x_vals = np.linspace(150, 200, 1000)
means = [171, 175]
variances = [7.1, 10]

plt.figure(figsize=(8, 5))
for mean, variance in zip(means, variances):
    y_vals = normal_density(mean, variance, x_vals)
    plt.plot(x_vals, y_vals, label=f"Mean={mean}, Variance={variance}")
plt.legend()
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Normal Distributions")
plt.show()

def integration(mean, variance, a, b):
    """Numerically integrates the normal density function between a and b."""
    result, _ = quad(lambda x: normal_density(mean, variance, x), a, b)
    return result

# Probability of a male's height being between 162cm and 190cm
prob = integration(171, 7.1, 162, 190)
print(f"Probability of height between 162cm and 190cm: {prob:.4f}")

#Using numerical integration, we found that the probability of a randomly chosen male having a height between 162 cm and 190 cm is approximately 0.9736. This means that about 97.36% of males fall within this height range. Since the normal distribution is centered around 171 cm with a small variance, most values are concentrated around this region, making it highly likely for a random male to be within this interval.
```
