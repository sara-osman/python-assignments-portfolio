```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')

# (a) 
f = sp.exp(x) * sp.sin(y) + y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
print("(a) Partial derivatives of f:")
print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)

# (b)
g = x**2 * y + x * y**2
grad_g = [sp.diff(g, x), sp.diff(g, y)]
grad_g_magnitude = sp.sqrt(grad_g[0]**2 + grad_g[1]**2).subs({x: 1, y: -1})
print("\n(b) Gradient of g and its magnitude at (1, -1):")
print("∇g =", grad_g)
print("|∇g| =", grad_g_magnitude)

# (c) Define h(x, y) and compute second partial derivatives
h = sp.ln(x**2 + y**2)
d2h_dx2 = sp.diff(h, x, x)
d2h_dy2 = sp.diff(h, y, y)
d2h_dxdy = sp.diff(h, x, y)
print("\n(c) Second partial derivatives of h:")
print("∂²h/∂x² =", d2h_dx2)
print("∂²h/∂y² =", d2h_dy2)
print("∂²h/∂x∂y =", d2h_dxdy)

# (d) Contour plot for j(x, y) = x^3 - 3xy + y^3
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.figure(figsize=(6,5))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# (e) Contour plot for another function: k(x, y) = sin(x)cos(y)
k = sp.sin(x) * sp.cos(y)
k_func = sp.lambdify((x, y), k, 'numpy')

Z_k = k_func(X, Y)

plt.figure(figsize=(6,5))
plt.contourf(X, Y, Z_k, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = \sin(x) \cos(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
